```python
import onnx
import numpy as np
from PIL import Image
import os

def convert_onnx_to_tflite(model_path, output_path, device_name="Samsung Galaxy S24 (Family)"):
    try:
        # Load ONNX model
        print(f"Loading ONNX model from {model_path}")
        onnx_model = onnx.load(model_path)

        # Get input shape from model
        input_shape = None
        for input in onnx_model.graph.input:
            shape = [dim.dim_value for dim in input.type.tensor_type.shape.dim]
            if len(shape) == 4:
                if shape[0] == 0 or shape[0] is None:
                    shape[0] = 1
                input_shape = tuple(shape)
                break

        if input_shape is None:
            input_shape = (1, 224, 224, 3)  # Default shape
        print(f"Using input shape: {input_shape}")

        # Upload model to hub first
        print("Uploading model to hub...")
        model_id = hub.upload_model(model_path)

        # Submit compile job with TFLite target
        print("Submitting compilation job...")
        compile_job = hub.submit_compile_job(
            model=model_id,
            device=hub.Device(device_name),
            input_specs={"input": input_shape},
            options="--target_runtime tflite"
        )
        assert isinstance(compile_job, hub.CompileJob)

        print("Waiting for compile job completion...")
        target_model = compile_job.get_target_model()
        if target_model is None:
            raise Exception("Compilation failed - target model is None")

        # Submit profile job
        print("Submitting profile job...")
        profile_job = hub.submit_profile_job(
            model=target_model,
            device=hub.Device(device_name)
        )
        assert isinstance(profile_job, hub.ProfileJob)
        print("Profile job submitted successfully")

        # Save the model
        print(f"Downloading model to {output_path}...")
        target_model.download(output_path)
        print(f"Model successfully converted and saved to: {output_path}")

        return True, target_model

    except Exception as e:
        print(f"Error during conversion: {str(e)}")
        import traceback
        traceback.print_exc()
        return False, None

def run_inference(model, image_path, device_name="Samsung Galaxy S24 (Family)"):
    try:
        # Load and preprocess image
        print(f"Loading and preprocessing image: {image_path}")
        image = Image.open(image_path).resize((224, 224))

        # Convert to array and apply MobileNetV3 preprocessing
        input_array = np.array(image, dtype=np.float32)
        input_array = (input_array / 127.5) - 1  # Scale to [-1, 1]

        # Handle grayscale images
        if input_array.ndim == 2:
            input_array = np.stack([input_array] * 3, axis=-1)

        # Add batch dimension (maintain NHWC format)
        input_array = np.expand_dims(input_array, axis=0)
        print(f"Preprocessed image shape: {input_array.shape}")

        # Submit inference job
        print("Submitting inference job...")
        inference_job = hub.submit_inference_job(
            model=model,
            device=hub.Device(device_name),
            inputs={"input": [input_array]}
        )

        return inference_job.download_output_data()

    except Exception as e:
        print(f"Error during inference: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

if __name__ == "__main__":
    # Paths
    model_path = '/content/drive/MyDrive/CS471_AI/FinalProject/MangoDataset_Balanced/mango_model.onnx'
    output_path = 'mango_ripeness_model.tflite'

    # Categories for mango ripeness (match the training order)
    categories = ['OverRipe', 'Ripe', 'Unripe', 'VeryRipe']  # Updated to include VeryRipe

    # Convert model
    print("Starting model conversion...")
    success, target_model = convert_onnx_to_tflite(model_path, output_path)

    if success:
        print("\nConversion successful!")

        # Run inference if you have a test image
        test_image_path = "https://drive.google.com/file/d/1_qpZ4HgQUqF6KfHXINg0-QfIcLNh-Mwy/view?usp=share_link"  # link test image path
        if os.path.exists(test_image_path):
            print("\nRunning inference on test image...")
            output_data = run_inference(target_model, test_image_path)

            if output_data:
                # Process outputs
                output_name = list(output_data.keys())[0]
                logits = output_data[output_name][0]

                # Apply softmax for probabilities
                probabilities = np.exp(logits - np.max(logits))
                probabilities = probabilities / np.sum(probabilities)

                # Print predictions
                print("\nMango Ripeness Predictions:")
                for i, prob in enumerate(probabilities):
                    print(f"{categories[i]:10s}: {prob:>6.1%}")
        else:
            print(f"Test image not found at: {test_image_path}")
    else:
        print("Conversion failed. Check the error messages above.")
```
